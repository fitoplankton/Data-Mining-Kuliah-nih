Jelaskan makna dataset. Jelaskan pula makna data training dan data testing

Dataset merupakan sebuah kumpulan data yang membentuk sebuah himpunan data. Biasanya, dalam data set terdapat baris dan kolom, pada kolom terdapat  
atribut/feature/dimension dan class/label/target. Nama lain dari data adalah record/object/sample/tuple. dan terdapat 2 tipe data, yaitu
numerik dan nominal

Data training adalah penggunaan beberapa data yang terdapat pada dataset bertujuan untuk melatih algoritma dalam mencari model yang sesuai, 
biasanya digunakan pada data yang akan digunakan dalam metode machine learning

Data testing adalah penggunaan beberapa data yang terdapat pada data set bertujuan untuk melakukan test atau pengujian terhadap model atau metode yang digunakan untuk mengetahui berapa tingkat performa model yang kita gunakan

Jelaskan proses penyusunan dataset
1. Aggregation(Aggregasi) : menyatukan 2 atribut atau lebih menjadi 1 atribut(data reduction), mengubah perubahan cakupan atau skala yang lebih tinggi
    (high-level) daripada yang lebih rendah(low-level)

2. Sampling(sampel): pengambilan beberapa data. 
    a. Sampling without replacement : penghapusan data yg digunakan dari populasi

    b. Sampling with replacement : tidak menghapus data dari populasi, objek yang sama dapat digunakan lebih dari 1 kali

    c. Stratified sampling : memisahkan data menjadi beberapa partisi, lalu dilakuka pengambilan acak dari partisi tersebut
    
    d. Progressive samplinge: digunakan ketika ukuran sample yang tepat cukup sulit ditentukan. Pendekatan inisample yang kecil, kemudian meningkatkan ukuran sample sampai sample dengan ukuran yang cukup diperoleh. Setiap item yang dipilih akan dihapus dari populasi

3. Dimensionality Reduction (reduksi dimensionalitas): mereduksi atau mengurangi atribut data yang digunakan. Untuk mengurangi noise

4. Feature Subset Selection : cara lain untuk mereduksi dimensi data dengan cara menghapuskan beberapa subset
    a. Redundant features : menghapus beberapa data yg terduplikat

    b. Irrelevant features : menghapus data yang kurang berpengaruh

5. Feature Creation : menambah atribut baru

6. Discretization and Binarization : melakukan transformasi dari data kontinyu menjadi data diskrit(ordinal) maupun biner

7. Attribute Transformation : memetakan kumpulan nilai dari atribut yang ada. Sehingga nilai lama dapat diidentifikasikan menggunakan nilai yang baru
