```python
import pandas as pd
```

```python
df = pd.read_csv("iris(1).csv")
df
```

```python
attribute_list = ["sepal.length", "sepal.width", "petal.length", "petal.width"]
```

```python
for a in attribute_list:
  min =  df[a].min()
  max = df[a].max()
  df[a] = (df[a] - min) / (max-min)

df
```

```python
class_iris = df["variety"].unique()
centroid = {}
for kelas_iris in class_iris:
  data_baru = df[df["variety"] == kelas_iris]
  centroid[kelas_iris] = {}
  for a in attribute_list :
    centroid[kelas_iris][a] = data_baru[a].mean()
    # print(kelas_iris, a)

centroid
```

```python
iris_test = pd.read_csv("iristesting.csv")
iris_test
```

```python
for a in attribute_list:
  min =  iris_test[a].min()
  max = iris_test[a].max()
  iris_test[a] = (iris_test[a] - min) / (max-min)

iris_test
```

```python
for kelas_iris in class_iris:
  jarak = 0
  for a in attribute_list:
    jarak += (iris_test[a] - centroid[kelas_iris][a])**2
  jarak = jarak**(1/2)
  iris_test[kelas_iris] = jarak

iris_test
```

```python
hasil_df = iris_test[class_iris]
# hasil_df
# hasil_df.min()
series_hasil =hasil_df.idxmin(axis=1)
series_hasil
```

```python
benar = (iris_test['variety'] == series_hasil).sum()

benar
```

```python
total = len(iris_test)
akurasi = (benar/total)*100
print("akurasinya = ", akurasi, "%")
```
